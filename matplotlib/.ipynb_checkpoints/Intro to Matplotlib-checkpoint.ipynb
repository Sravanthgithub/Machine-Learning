### Matplotlib Example Workflow

```python
import matplotlib.pyplot as plt
```

```python
#prep of data
x = [1,2,3,4]
y = [11,22,33,44]

#setup plot
fig, ax = plt.subplots(figsize=(10,10))

#plot data
ax.plot(x,y)

#customize plot
ax.set(title='Simple plot',xlabel='x-axis',ylabel='y-axis');
```

### Making figs with numpy arrays

```python
import numpy as np
```

```python
x = np.linspace(0,10,100)
```

```python
#line plot
fig,ax = plt.subplots()
ax.plot(x,x**2);
```

```python
#scatter plot
fig,ax = plt.subplots()
ax.scatter(x,np.sin(x));
```

```python
prices = {"Almond":10,
         "Peanut": 8,
         "Cashew": 12}
fig,ax = plt.subplots()
ax.bar(prices.keys(),prices.values());
```

```python
fig,ax = plt.subplots()
ax.barh(list(prices.keys()),list(prices.values()));
```

```python
x = np.random.randn(1000)
fig,ax = plt.subplots()
ax.hist(x);
```

```python
#subplots option 1
fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows=2,
                                          ncols=2,
                                          figsize=(10,5))

ax1.plot(x,x/2);
ax2.plot(x,x/4);
ax3.plot(x,x);
ax4.plot(x,x**2);
```

```python
# subplot option 2
fig, ax = plt.subplots(nrows=2,
                                          ncols=2,
                                          figsize=(10,5))
ax[0,0].plot(x,x/2);
ax[0,1].plot(x,x/4);
ax[1,0].plot(x,x);
ax[1,1].plot(x,x**2);
```
