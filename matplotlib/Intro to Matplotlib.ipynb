```python
x = np.random.randn(1000)
fig,ax = plt.subplots()
ax.hist(x);
```

```python
#subplots option 1
fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows=2,
                                          ncols=2,
                                          figsize=(10,5))

ax1.plot(x,x/2);
ax2.plot(x,x/4);
ax3.plot(x,x);
ax4.plot(x,x**2);
```

```python
# subplot option 2
fig, ax = plt.subplots(nrows=2,
                                          ncols=2,
                                          figsize=(10,5))
ax[0,0].plot(x,x/2);
ax[0,1].plot(x,x/4);
ax[1,0].plot(x,x);
ax[1,1].plot(x,x**2);
```

```python
over_50 = hd[hd['age']>50]
over_50.head()
```

```python
over_50.plot(kind='scatter',
             x='age',
             y='chol',
             c='target');
```

```python
fix, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter',
             x='age',
             y='chol',
             c='target',
             ax=ax);
ax.set_xlim([45,100]);
```

```python
#OO method from scratch
fix, ax = plt.subplots(figsize=(10,6))

#plot the data
scatter = ax.scatter(x=over_50['age'],
                     y=over_50['chol'],
                     c=over_50['target'])

#customise the plot
ax.set(title='Heart disease and cholestrol levels',
      xlabel='age',
      ylabel='cholestrol')
#add a legend
ax.legend(*scatter.legend_elements(), title='Target');
```

```python
over_50.head()
```

```python
fig, (ax0,ax1) = plt.subplots(nrows=2,ncols=1,
                              figsize=(10,10))


scatter = ax0.scatter(x=over_50['age'],
                     y=over_50['chol'],
                     c=over_50['target'])

ax0.set(title='Heart disease and cholestrol levels',
      xlabel='age',
      ylabel='cholestrol')

ax0.axhline(y=over_50['chol'].mean(),linestyle='--');

scatter = ax1.scatter(x=over_50['age'],
                     y=over_50['thalach'],
                     c=over_50['target'])

ax1.legend(*scatter.legend_elements(),title='Target')
ax1.axhline(y=over_50['thalach'].mean(),linestyle='--');


fig.suptitle("Heart Disease Analysis",fontsize=16,fontweight='bold');
```

```python
fig, (ax0,ax1) = plt.subplots(nrows=2,ncols=1,
                              figsize=(10,10))


scatter = ax0.scatter(x=over_50['age'],
                     y=over_50['chol'],
                     c=over_50['target'],
                     cmap='winter')

ax0.set(title='Heart disease and cholestrol levels',
      xlabel='age',
      ylabel='cholestrol',xlim=(50,80))
#changind x axis limits


ax0.axhline(y=over_50['chol'].mean(),linestyle='--');

scatter = ax1.scatter(x=over_50['age'],
                     y=over_50['thalach'],
                     c=over_50['target'],
                     cmap='summer')

ax1.set(title='Heart disease and max hear rate',
      xlabel='age',
      ylabel='max heart rate',xlim=(50,80),
       ylim=(60,200))

ax1.legend(*scatter.legend_elements(),title='Target')

ax1.axhline(y=over_50['thalach'].mean(),linestyle='--');


fig.suptitle("Heart Disease Analysis",fontsize=16,fontweight='bold');
```
